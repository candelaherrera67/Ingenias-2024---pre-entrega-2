```python
#Realizo otro gráfico, como mi variable va a estar escrita entonces no me conviene usar este gráfico. En dicho gráfico se puede observar la relación entre el ingreso y gastos por empresa
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='tipo_estudio', y='empresa', data=df, palette='viridis')
plt.title('Relación entre Ingreso y Puntuación de Gastos')
plt.xlabel('tipo_estudio')
plt.ylabel('Puntuación de Gastos')
plt.legend(title='ingreso y empresas')
plt.show()
```
